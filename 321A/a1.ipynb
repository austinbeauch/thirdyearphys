```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
```

```python
g = 9.81
def dydt(initial, t):
    h, v = initial
    return [v, -g]
```

```python
h0 = 32000
v0 = 0
t0 = 0 
tf = 100
step = 10000
m = 70

y0 = [h0, v0]

times = np.linspace(t0, tf, step)
```

```python
sol = odeint(dydt, y0, times)
```

```python
plt.plot(times, sol[:, 0])
plt.xlabel("Time (s)"); plt.ylabel("Height (m)"); plt.title("Air Firctionless Parachutist")
```

```python
ground = round(times[np.argwhere(sol[:, 0]<=0)[0][0]])
print(f"The parachutist hit the ground at {int(ground)}s.")
```

```python
c = 0.5
def dydt_2(initial, t):
    h, v = initial
    return [v, -(c/m) * v *np.abs(v) - g]
```

```python
tf2 = 1000
times2 = np.linspace(t0, tf2, step)
```

```python
sol = odeint(dydt_2, y0, times2)
```

```python
plt.plot(times2, sol[:, 0])
plt.xlabel("Time (s)"); plt.ylabel("Height (m)"); plt.title("Air Resistance Parachutist")
```

```python
ground = round(times2[np.argwhere(sol[:, 0]<=0)[0][0]])
print(f"The parachutist hit the ground at {int(ground)}s.")
```
