```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
domain = np.linspace(0, 1e-9, 100)
hbar = 1.05*10**-34
```

```python
def magsq(z):
    return z.real**2 + z.imag**2

def psi(x, t, a=1e-9, m=9.1*10**-31):
    const = 2*np.sqrt(420)/(np.pi**3 * np.sqrt(a)) 
    
    wave = 0
    for n in range(1,10):
        f = (2 + (-1)**n) / n**3
        s = np.sin(n * np.pi * x / a)
        e = np.exp(-1j*n**2 * np.pi**2 * hbar * t / (2*m*a**2))
        wave += f*s*e
    return magsq(const*wave)
```

```python
plt.figure(figsize=(8,5))
for t in [0, 0.1e-15, 0.8e-15, 1.8e-15]:
    plt.plot(domain, psi(domain, t), label=f"t={t}s")
plt.legend()
plt.title("Wavefunction Evolving in Time")
plt.xlabel("Position (m)")
plt.ylabel("$\Psi (x,y)$")
plt.show()
```
