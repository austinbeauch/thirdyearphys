```python
from numpy import *
from matplotlib import pyplot as plt
```

```python
def magsq(z):
    return z.real**2 + z.imag**2

def psi_200(x, y, z, Z=1, a0=1):
    r = sqrt(x**2 + y**2 + z**2)
    return 1/sqrt(pi) * (Z/(2*a0))**(3/2) * (1 - (Z*r/(2*a0))) * exp(-Z*r/(2*a0))

def psi_210(x, y, z, Z=1, a0=1):
    r = sqrt(x**2 + y**2 + z**2)
    return 1/(2*sqrt(pi)) * (Z/(2*a0))**(3/2) * (Z*r/a0) * exp(-Z*r/(2*a0)) * (z/r)

def psi(x,y,z, alpha=0):
    return magsq(sqrt(4/5)*psi_200(x,y,z) + exp(1j*alpha)*sqrt(1/5)*psi_210(x,y,z))
```

```python
dom = linspace(-10, 10, 1000)
x, z = meshgrid(dom, dom)
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
ax.pcolormesh(x, z,  magsq(psi_200(x, 0, z)), cmap='gray',  vmin=0, vmax=0.0014)
ax = plt.subplot(1,2,2)
ax.pcolormesh(x, z,  magsq(psi_210(x, 0, z)), cmap='gray',  vmin=0, vmax=0.0014)
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i, alpha in enumerate([0, pi/4, pi/3, pi/2]):
    ax = plt.subplot(2,2,i+1)
    ax.pcolormesh(x, z, psi(x, 0, z, alpha), cmap='gray',  vmin=0, vmax=0.0014)
```
